# Multivariate Analysis

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

```python
data = pd.read_csv("car_price_dataset.csv")
```

```python
data.head()
```

> Printing first 5 rows

#### NUMERICAL-NUMERICAL

#### Scatter plot (Price vs mileage)

```python
sns.scatterplot(data=data,x="Mileage",y="Price")
```

>There is a negative correlation between Mileage and Price. As Mileage increases, Price tends to decrease.
> This makes sense because cars with higher mileage are usually older or more used, leading to lower resale values.

### Categorical vs numerical

#### Barplot (fuel_type vs engine_size)

```python
plt.figure(figsize=(8, 5))
sns.barplot(data=data,x="Fuel_Type",y="Engine_Size",hue="Fuel_Type", palette="viridis", legend=False)
plt.show()
```

> Mean Engine_Size for every fuel type is around 3....

#### Bar plot (Brand vs Engine size)

```python
plt.figure(figsize=(10, 5))
sns.barplot(data=data,x="Brand",y="Engine_Size",hue="Brand", palette="viridis", legend=False)
plt.show()
```

> Mercedes and Kia cars are having higher mean of Enginer_Size, where as other cars have close to 3.0

#### Boxplot (Brand vs Doors)

```python
plt.figure(figsize=(10, 5))
sns.boxplot(data=data,x='Brand',y='Doors',hue='Brand',palette='viridis',legend=True)
```

> Most brands have a median of 3-4 doors, suggesting that these cars are the most common.
> Some brands (like Kia, Toyota, and Hyundai) seem to have a wider range, meaning they produce cars with both higher and lower door counts.

#### Box plot (brand vs door, hue=owner_count)

```python
plt.figure(figsize=(20, 5))
sns.boxplot(data=data,x='Brand',y='Doors',hue='Owner_Count',legend=True)
```

> Most cars have 4 doors, but some brands (like Kia, Ford, and Volkswagen) also have models with 2 or 5 doors.
Owner Count Variation:
Cars with fewer owners (lighter colors) appear to be more common across brands.
Darker colors (higher owner counts) are scattered, indicating some older, resold vehicles.

#### Box plot (brand vs year)

```python
plt.figure(figsize=(10,5))
sns.boxplot(data=data,x="Brand", y="Year",hue='Brand',palette='viridis',legend=False)
plt.title('')
plt.xticks(rotation=80)
plt.show()
```

>the median year is b/w 2010 and 2015, no outliears, wide range represents the longer years in the market.

#### barplot (Brand vs Price , hue=Doors)

```python
plt.figure(figsize=(10,5))
sns.barplot(data=data,x="Brand", y="Price",hue='Doors',palette='viridis',legend=True)
plt.title('Brand vs Price')
plt.xticks(rotation=80)
plt.show()
```

> Prices Are Relatively Close Across Door Categories:
The bars for each brand have small variations, meaning the number of doors does not significantly affect price for most brands.
However, slight differences are noticeable for some brands.

>Higher Average Price for Certain Brands:
Brands like Mercedes, Audi, and BMW have consistently higher prices.
More affordable brands like Kia, Hyundai, and Chevrolet show lower average prices.

>No Clear Increasing Trend With More Doors:
In some brands, 4 or 5-door models aren't necessarily more expensive than 2 or 3-door models.
This suggests that factors other than door count (e.g., features, performance, and brand positioning) influence price.

#### Scatter plot (Doors vs price) 

```python
plt.figure(figsize=(10,5))
sns.scatterplot(data=data,x='Doors',y='Price')
```

> Distinct Clusters:
Cars are grouped at 2, 3, 4, and 5 doors without any in-between values.
This suggests that cars come in these standard door.

>Price Range Across Doors:
Prices are spread across all door configurations, meaning no clear direct relationship between the number of doors and price.
However, cars with 2 and 3 doors may belong to sportier or compact models, while 4 and 5 doors might include sedans, SUVs, or larger vehicles.

#### Scatter plot (Price vs brand)

```python
plt.figure(figsize=(10,5))
sns.scatterplot(data=data,x='Brand',y='Price')
```

>Price Distribution:
Each brand has a wide price range, meaning there are both budget-friendly and high-end models.
The prices appear to be mostly between $2,000 and $18,000.
Luxury vs. Economy Brands:

>Brands like Mercedes, Audi, and BMW likely have higher prices on average.
Brands like Kia, Hyundai, and Chevrolet may have more affordable options.
Data Density:

> Some brands have more cars in the dataset than others, based on the concentration of dots.
